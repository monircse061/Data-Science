```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
df = pd.read_csv('G:/Old pc important files/Desktop/thesis 2021/employee attrition dataset/employee_attrition.csv')
```

```python
df.head()
```

# Column Value Replace

```python
df['Attrition'].replace({'Yes':1,'No':0}, inplace = True)
```

```python
df.head()
```

```python
df['Attrition'].value_counts()
```

```python
df.isnull().sum()
```

# Data Preprocessing

```python
dummies = pd.get_dummies(df[['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','Over18','OverTime']])
```

```python
marged_df = pd.concat([df,dummies] , axis='columns')
```

```python
df = marged_df.drop(['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','Over18','OverTime'], axis = 1)
```

```python
df.head()
```

# Split target and input variable

```python
x = df.drop(['Attrition'],axis = 1)
y = df['Attrition']
```

# Univariate Selection 

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
fit_feature = SelectKBest(score_func = f_classif)
```

```python
fit_feature.fit(x,y)
```

```python
score_col = pd.DataFrame(fit_feature.scores_,columns = ['Score value'])
```

```python
score_col
```

```python
Names = pd.DataFrame(x.columns)
top_features = pd.concat([Names, score_col], axis = 1)
```

```python
top_features
```

```python
top_features.nlargest(55,'Score value')
```

```python
columns = ['TrainingTimesLastYear','WorkLifeBalance','EducationField_Technical Degree','BusinessTravel_Non-Travel','DistanceFromHome','JobRole_Healthcare Representative','Department_Sales','JobRole_Manufacturing Director','JobRole_Manager','Department_Research & Development','MaritalStatus_Divorced','JobRole_Research Director','JobRole_Sales Executive','Gender_Male','Gender_Female','Education','EducationField_Other','Department_Human Resources','MonthlyRate','PercentSalaryHike','EmployeeNumber','HourlyRate','PerformanceRating','JobRole_Research Scientist']
```

```python
x = x.drop(columns, axis =1)
```

```python
x
```

```python
seed = 7
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.2,random_state=7 )
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rnf = RandomForestClassifier()
```

```python
rnf.fit(xtrain,ytrain)
```

```python
# Accuracy on Training 
predictions_train = rnf.predict(xtrain)
print("\nTraining Accuracy : ", accuracy_score(ytrain,predictions_train))
# Accuracy on Testing
predictions_test = rnf.predict(xtest)
print("\nTesting Accuracy : ", accuracy_score(ytest,predictions_test))
```

# SVM Model

```python
svm_model = SVC(gamma = 'auto')
svm_model.fit(xtrain,ytrain)
```

```python
svm_model.score(x,y)
```

```python
# Accuracy on Training 
predictions_train = svm_model.predict(xtrain)
print("\nTraining Accuracy : ", accuracy_score(ytrain,predictions_train))
# Accuracy on Testing
predictions_test = svm_model.predict(xtest)
print("\nTesting Accuracy : ", accuracy_score(ytest,predictions_test))
```
