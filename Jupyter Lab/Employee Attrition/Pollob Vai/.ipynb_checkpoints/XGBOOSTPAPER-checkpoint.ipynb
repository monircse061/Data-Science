```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
import seaborn as sns
```

```python
data= pd.read_csv("G:/Datasets/Dataset_for_Classification.csv")
data.head(10)
```

```python
label_encoder =  preprocessing.LabelEncoder()
data['BusinessTravel'] = label_encoder.fit_transform(data['BusinessTravel'])
data['Attrition'] = label_encoder.fit_transform(data['Attrition'])
data['Department'] = label_encoder.fit_transform(data['Department'])
data['EducationField'] = label_encoder.fit_transform(data['EducationField'])
data['Gender'] = label_encoder.fit_transform(data['Gender'])
data['JobRole'] = label_encoder.fit_transform(data['JobRole'])
data['MaritalStatus'] = label_encoder.fit_transform(data['MaritalStatus'])
data['OverTime'] = label_encoder.fit_transform(data['OverTime'])
```

```python
seed = 5
train_data, test_data = train_test_split(data, test_size=0.2, random_state=seed)
```

```python
train_x = train_data.drop(columns=['Attrition'],axis=1)
train_y = train_data['Attrition']
```

```python
std_train_x1=train_x
#std_train_x1=preprocessing.scale(train_x)
```

```python
test_x = test_data.drop(columns=['Attrition'],axis=1)
test_y = test_data['Attrition']
```

```python
std_test_x1=test_x
#std_test_x1=preprocessing.scale(test_x)
```

```python
from xgboost import XGBClassifier
classifier=XGBClassifier()
classifier.fit(std_train_x1,train_y)
predictions_train =classifier.predict(std_train_x1)
acc=accuracy_score(train_y, predictions_train)
print('\nTraining Accuracy :', acc)
predictions_test = classifier.predict(std_test_x1)
accc=accuracy_score(test_y, predictions_test)
print('\nTesting Accuracy :', accc)
```
